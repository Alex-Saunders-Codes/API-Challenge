### Hotel Map
* Store into variable named `hotel_df`.
* Add a "Hotel Name" column to the DataFrame.
* Set parameters to search for hotels with 5000 meters.
* Hit the Google Places API for each city's coordinates.
* Store the first Hotel result into the DataFrame.
* Plot markers on top of the heatmap.

```python
hotel_df = ideal.copy()
hotel_df.head()
```

```python
searchurl=f'https://maps.googleapis.com/maps/api/place/nearbysearch/json?input=Hotel&inputtype=textquery&fields=name,rating&locationbias=circle:radius@{hotel_df.iloc[0]["lat"]},{hotel_df.iloc[0]["lng"]}&key={g_key}'
```

```python
searchurl2 = f'https://maps.googleapis.com/maps/api/place/nearbysearch/json?location={hotel_df.iloc[0]["lat"]},{hotel_df.iloc[0]["lng"]}&radius=5000&type=hotel&key={g_key}&rankby=prominence'
```

```python
response = requests.get(searchurl2).json()
response['results'][1]['rating']
```

```python
hotel_names = []
hotel_ratings = []
for location in hotel_df.itertuples(index=False):
    vacay_not_found = True
    #perform search queries of hotels within 5,000 meters of my ideal locations.
    searchurl = f'https://maps.googleapis.com/maps/api/place/nearbysearch/json?location={location.lat},{location.lng}&radius=5000&type=hotel&key={g_key}&rankby=prominence'
    response = requests.get(searchurl).json()['results']
    iterator = 0 
    
    try:
        while vacay_not_found:
            if ('name' in response[iterator].keys()) and ('rating' in response[iterator].keys())
                hotel_names.append(response[]['name'])
                hotel_ratings.append(response['rating'])
                vacay_not_found = False
            else: 
                iterator += 1
    except Exception as e:
        print(e)
hotel_df['Hotel Name'] = hotel_names
hotel_df['Hotel Ratings'] = hotel_ratings
hotel_df.head()
```
